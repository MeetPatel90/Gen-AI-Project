```python
# ============================================================================
# HYBRID TB–SCORE GFLOWNET - FINAL IMPLEMENTATION WITH COMPARISON EXPERIMENT
# ============================================================================

import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.distributions import Normal
import matplotlib.pyplot as plt
from collections import deque
import random

# Use GPU if available
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Running on device: {device}")

# ============================================================================
# SECTION 1: THE PHYSICAL SYSTEM
# ============================================================================

class DoubleWellPotential:
    """Potential Energy Surface U(x) = (x^2 - 1)^2"""
    def energy(self, x): return (x**2 - 1)**2
    def force(self, x): return -4 * x * (x**2 - 1)

class MDSimulator:
    """Overdamped Langevin Dynamics Simulator"""
    def __init__(self, potential, D=0.1, dt=0.01, device='cpu'):
        self.potential, self.D, self.dt, self.device = potential, D, dt, device

    def step(self, x, bias=None):
        drift = self.potential.force(x) * self.dt
        if bias is not None:
            drift += bias * self.dt
        noise = torch.randn_like(x) * np.sqrt(2 * self.D * self.dt)
        return x + drift + noise

    def rollout(self, x0, T, bias_fn=None):
        trajectory = [x0.clone()]
        x = x0.clone()
        for t in range(T):
            bias = bias_fn(x, t) if bias_fn is not None else None
            x = self.step(x, bias)
            trajectory.append(x.clone())
        return torch.stack(trajectory)

class OMActionCalculator:
    """Calculates Onsager-Machlup Action"""
    def __init__(self, potential, D=0.1, dt=0.01):
        self.potential, self.D, self.dt = potential, D, dt

    def compute_action(self, trajectory):
        x_prev, x_curr = trajectory[:-1], trajectory[1:]
        force = self.potential.force(x_prev)
        residuals = x_curr - x_prev - force * self.dt
        action = (residuals**2).sum(dim=-1) / (4 * self.D * self.dt)
        return action.sum()

# ============================================================================
# SECTION 2: THE GFLOWNET COMPONENTS
# ============================================================================

class BiasNetwork(nn.Module):
    """The forward policy's bias force b_θ(x)"""
    def __init__(self, input_dim=1, hidden_dim=64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim), nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim), nn.ReLU(),
            nn.Linear(hidden_dim, input_dim)
        )
    def forward(self, x): return self.net(x)

class BackwardNetwork(nn.Module):
    """The backward policy's bias force b_φ(x)"""
    def __init__(self, input_dim=1, hidden_dim=64):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim), nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim), nn.ReLU(),
            nn.Linear(hidden_dim, input_dim)
        )
    def forward(self, x): return self.net(x)

class ReplayBuffer:
    """Experience replay buffer"""
    def __init__(self, capacity=10000):
        self.buffer = deque(maxlen=capacity)
    def add(self, trajectory):
        self.buffer.append(trajectory.detach().cpu())
    def sample(self, batch_size, device='cpu'):
        if len(self.buffer) < batch_size:
            return []
        batch = random.sample(self.buffer, batch_size)
        return [item.to(device) for item in batch]
    def __len__(self):
        return len(self.buffer)

# ============================================================================
# SECTION 3: THE CORRECT HYBRID LOSS
# ============================================================================

class TrajectoryBalanceLoss:
    def __init__(self, potential, bias_net, backward_net, D, dt):
        self.potential = potential
        self.bias_net = bias_net
        self.backward_net = backward_net
        self.D, self.dt = D, dt
        self.log_Z = nn.Parameter(torch.tensor(0.0))
        self.om_calc = OMActionCalculator(potential, D, dt)

    def calculate_loss(self, trajectory):
        x_prev, x_curr = trajectory[:-1], trajectory[1:]
        var = 2 * self.D * self.dt

        # Forward probability log P_F(τ)
        force_f = self.potential.force(x_prev)
        bias_f = self.bias_net(x_prev)
        mean_f = x_prev + (force_f + bias_f) * self.dt
        log_pF = Normal(mean_f, np.sqrt(var)).log_prob(x_curr).sum()

        # Backward probability log P_B(τ)
        # The backward dynamics are also biased by the physical force
        force_b = self.potential.force(x_curr)
        bias_b = self.backward_net(x_curr)
        mean_b = x_curr - (force_b + bias_b) * self.dt # Note the sign change for time reversal
        log_pB = Normal(mean_b, np.sqrt(var)).log_prob(x_prev).sum()

        # Log Reward log R(τ) = -S(τ) + log I(x_T)
        action = self.om_calc.compute_action(trajectory)
        x_final = trajectory[-1]
        log_indicator = -((x_final - 1.0)**2).sum() / (2 * 0.1**2) # Soft indicator for region B
        log_R = -action + log_indicator

        tb_error = self.log_Z + log_pF - log_R - log_pB
        return tb_error**2

# ============================================================================
# SECTION 4: THE TRAINER
# ============================================================================

class Trainer:
    def __init__(self, potential, D=0.1, dt=0.01, T=200, device='cpu'):
        self.potential = potential
        self.T = T
        self.device = device

        self.bias_net = BiasNetwork(input_dim=1).to(device)
        self.backward_net = BackwardNetwork(input_dim=1).to(device)

        self.loss_fn = TrajectoryBalanceLoss(potential, self.bias_net, self.backward_net, D, dt)

        params = (list(self.bias_net.parameters()) +
                  list(self.backward_net.parameters()) +
                  [self.loss_fn.log_Z])
        self.optimizer = optim.Adam(params, lr=1e-4)

        self.simulator = MDSimulator(potential, D, dt, device)
        self.buffer = ReplayBuffer()
        self.history = {'tb_loss': []}

    def train(self, n_iterations=100, explore_steps=64, train_steps=32, batch_size=16):
        print("Starting GFlowNet training...")
        for i in range(n_iterations):
            # --- Exploration Phase ---
            for _ in range(explore_steps):
                x0 = torch.tensor([-1.0], device=self.device)
                bias_fn = lambda x, t: self.bias_net(x)
                with torch.no_grad():
                    traj = self.simulator.rollout(x0, self.T, bias_fn)
                self.buffer.add(traj)

            # --- Training Phase ---
            if len(self.buffer) < batch_size: continue

            iter_losses = []
            for _ in range(train_steps):
                batch = self.buffer.sample(batch_size, device=self.device)

                loss_batch = 0
                for trajectory in batch:
                    loss_batch += self.loss_fn.calculate_loss(trajectory)

                loss_batch /= len(batch)

                self.optimizer.zero_grad()
                loss_batch.backward()
                self.optimizer.step()

                iter_losses.append(loss_batch.item())

            self.history['tb_loss'].append(np.mean(iter_losses))

            if (i + 1) % 10 == 0:
                print(f"Iter {i+1}/{n_iterations} | TB Loss: {self.history['tb_loss'][-1]:.3e}")

        print("Training complete!")

# ============================================================================
# SECTION 5: EXPERIMENT & VISUALIZATION
# ============================================================================

def sample_unbiased_reactive_paths(simulator, n_paths, max_len, start_x, target_thresh):
    """Runs brute-force MD to find successful A->B transitions."""
    print(f"\nAttempting to find {n_paths} unbiased reactive paths...")
    reactive_paths = []
    max_trials = 50000
    for trial in range(max_trials):
        x0 = torch.tensor([start_x], device=simulator.device)
        with torch.no_grad():
            traj = simulator.rollout(x0, max_len)

        if traj[-1].item() > target_thresh:
            reactive_paths.append(traj.cpu())
            print(f"  > Success! Found reactive path #{len(reactive_paths)} on trial {trial+1}.")
            if len(reactive_paths) >= n_paths:
                break

    if not reactive_paths:
        print("  > FAILED to find any unbiased reactive paths in the given trials.")
    return reactive_paths

def plot_comparison(potential, unbiased_trajs, biased_trajs):
    """Plots both sets of trajectories on the potential and over time."""
    # --- Plot on Potential Energy Surface ---
    fig, ax = plt.subplots(1, 2, figsize=(18, 6))
    x_coords = torch.linspace(-2, 2, 200).unsqueeze(-1)
    U = potential.energy(x_coords).numpy()

    ax[0].plot(x_coords.numpy(), U, 'k-', label="Potential U(x)")

    for i, traj in enumerate(unbiased_trajs):
        label = "Unbiased MD" if i == 0 else None
        ax[0].plot(traj.numpy(), potential.energy(traj).numpy(), 'b-', alpha=0.6, label=label)

    for i, traj in enumerate(biased_trajs):
        label = "Biased GFlowNet" if i == 0 else None
        ax[0].plot(traj.numpy(), potential.energy(traj).numpy(), 'r-', alpha=0.6, label=label)

    ax[0].set_title("Trajectories on the Potential Energy Surface")
    ax[0].set_xlabel("x"); ax[0].set_ylabel("U(x)"); ax[0].legend(); ax[0].grid(True)

    # --- Plot Position vs. Time ---
    if unbiased_trajs:
        ax[1].plot(unbiased_trajs[0].numpy(), 'b-', alpha=0.8, label="Unbiased MD")
    if biased_trajs:
        ax[1].plot(biased_trajs[0].numpy(), 'r-', alpha=0.8, label="Biased GFlowNet")

    ax[1].axhline(1.0, color='grey', linestyle='--', alpha=0.7)
    ax[1].axhline(-1.0, color='grey', linestyle='--', alpha=0.7)
    ax[1].set_title("Position vs. Time (Example Trajectory)")
    ax[1].set_xlabel("Time Step"); ax[1].set_ylabel("x"); ax[1].legend(); ax[1].grid(True)

    plt.suptitle("Comparison of Unbiased vs. GFlowNet-Biased Trajectories")
    plt.show()
```

```python
# ============================================================================
# HYBRID TB–SCORE GFLOWNET - FINAL, CORRECTED IMPLEMENTATION
# ============================================================================

import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.distributions import Normal
import matplotlib.pyplot as plt
from collections import deque
import random

# Use GPU if available
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Running on device: {device}")

# ============================================================================
# SECTION 1: THE PHYSICAL SYSTEM
# ============================================================================

class DoubleWellPotential:
    """Potential Energy Surface U(x) = (x^2 - 1)^2"""
    def energy(self, x): return (x**2 - 1)**2
    def force(self, x): return -4 * x * (x**2 - 1)

class MDSimulator:
    """Overdamped Langevin Dynamics Simulator"""
    def __init__(self, potential, D=0.1, dt=0.01, device='cpu'):
        self.potential, self.D, self.dt, self.device = potential, D, dt, device

    def step(self, x, bias=None):
        drift = self.potential.force(x) * self.dt
        if bias is not None:
            drift += bias * self.dt
        noise = torch.randn_like(x) * np.sqrt(2 * self.D * self.dt)
        return x + drift + noise

    def rollout(self, x0, T, bias_fn=None):
        trajectory = [x0.clone()]
        x = x0.clone()
        for t in range(T):
            bias = bias_fn(x, t) if bias_fn is not None else None
            x = self.step(x, bias)
            trajectory.append(x.clone())
        return torch.stack(trajectory)

class OMActionCalculator:
    """Calculates Onsager-Machlup Action and its gradient (Score)"""
    def __init__(self, potential, D=0.1, dt=0.01):
        self.potential, self.D, self.dt = potential, D, dt

    def compute_action(self, trajectory):
        x_prev, x_curr = trajectory[:-1], trajectory[1:]
        force = self.potential.force(x_prev)
        residuals = x_curr - x_prev - force * self.dt
        action = (residuals**2).sum(dim=-1) / (4 * self.D * self.dt)
        return action.sum()

    def compute_score_targets(self, trajectory):
        x_prev, x_curr = trajectory[:-1], trajectory[1:]
        force = self.potential.force(x_prev)
        r = x_curr - x_prev - force * self.dt

        score_interior = (r[1:] - r[:-1]) / (2 * self.D * self.dt)
        return score_interior

# ============================================================================
# SECTION 2: THE GFLOWNET COMPONENTS
# ============================================================================

class BiasNetwork(nn.Module):
    def __init__(self, input_dim=1, hidden_dim=64):
        super().__init__()
        self.net = nn.Sequential(nn.Linear(input_dim, hidden_dim), nn.ReLU(), nn.Linear(hidden_dim, hidden_dim), nn.ReLU(), nn.Linear(hidden_dim, input_dim))
    def forward(self, x): return self.net(x)

class BackwardNetwork(nn.Module):
    def __init__(self, input_dim=1, hidden_dim=64):
        super().__init__()
        self.net = nn.Sequential(nn.Linear(input_dim, hidden_dim), nn.ReLU(), nn.Linear(hidden_dim, hidden_dim), nn.ReLU(), nn.Linear(hidden_dim, input_dim))
    def forward(self, x): return self.net(x)

class ScoreNetwork(nn.Module):
    def __init__(self, input_dim=1, hidden_dim=64):
        super().__init__()
        self.net = nn.Sequential(nn.Linear(input_dim, hidden_dim), nn.ReLU(), nn.Linear(hidden_dim, hidden_dim), nn.ReLU(), nn.Linear(hidden_dim, input_dim))
    def forward(self, x): return self.net(x)

class ReplayBuffer:
    def __init__(self, capacity=10000):
        self.buffer = deque(maxlen=capacity)
    def add(self, trajectory):
        self.buffer.append(trajectory.detach().cpu())
    def sample(self, batch_size, device='cpu'):
        if len(self.buffer) < batch_size: return []
        batch = random.sample(self.buffer, batch_size)
        return [item.to(device) for item in batch]
    def __len__(self): return len(self.buffer)

# ============================================================================
# SECTION 3: THE CORRECTED HYBRID LOSS
# ============================================================================

class HybridLoss:
    def __init__(self, potential, bias_net, backward_net, score_net, D, dt, alpha):
        self.potential, self.bias_net, self.backward_net, self.score_net = potential, bias_net, backward_net, score_net
        self.D, self.dt, self.alpha = D, dt, alpha
        self.log_Z = nn.Parameter(torch.tensor(0.0))
        self.om_calc = OMActionCalculator(potential, D, dt)

    def calculate_losses(self, trajectory):
        T = trajectory.shape[0] - 1
        x_prev, x_curr = trajectory[:-1], trajectory[1:]
        var = 2 * self.D * self.dt
        std = np.sqrt(var)

        # --- 1. Trajectory Balance Loss (Normalized) ---
        force_f = self.potential.force(x_prev)
        bias_f = self.bias_net(x_prev)
        mean_f = x_prev + (force_f + bias_f) * self.dt
        log_pF = Normal(mean_f, std).log_prob(x_curr).sum() / T

        force_b = self.potential.force(x_curr)
        bias_b = self.backward_net(x_curr)
        mean_b = x_curr - (force_b + bias_b) * self.dt
        log_pB = Normal(mean_b, std).log_prob(x_prev).sum() / T

        action = self.om_calc.compute_action(trajectory)
        x_final = trajectory[-1]
        log_indicator = -((x_final - 1.0)**2).sum() / (2 * 0.5**2) # Loosened sigma
        log_R = (-action + log_indicator) / T

        tb_error = self.log_Z + log_pF - log_R - log_pB
        loss_tb = tb_error**2

        # --- 2. Score Regularization Loss ---
        loss_score = torch.tensor(0.0, device=trajectory.device)
        if self.alpha > 0 and T > 1:
            score_targets = self.om_calc.compute_score_targets(trajectory)
            score_preds = self.score_net(trajectory[1:-1])
            if score_preds.shape[0] > 0:
                loss_score = nn.functional.mse_loss(score_preds, score_targets)

        # --- 3. Total Hybrid Loss ---
        total_loss = loss_tb + self.alpha * loss_score

        return total_loss, loss_tb.detach(), loss_score.detach()

# ============================================================================
# SECTION 4: THE TRAINER
# ============================================================================

class Trainer:
    def __init__(self, potential, D=0.1, dt=0.01, T=200, alpha=0.1, device='cpu'):
        self.potential, self.T, self.device = potential, T, device

        self.bias_net = BiasNetwork().to(device)
        self.backward_net = BackwardNetwork().to(device)
        self.score_net = ScoreNetwork().to(device)

        self.loss_fn = HybridLoss(potential, self.bias_net, self.backward_net, self.score_net, D, dt, alpha)

        params = (list(self.bias_net.parameters()) + list(self.backward_net.parameters()) +
                  list(self.score_net.parameters()) + [self.loss_fn.log_Z])
        self.optimizer = optim.Adam(params, lr=1e-4)

        self.simulator = MDSimulator(potential, D, dt, device)
        self.buffer = ReplayBuffer()
        self.history = {'total': [], 'tb': [], 'score': []}

    def add_to_buffer(self, trajectory):
        """Helper to trim long trajectories before adding to buffer."""
        if trajectory.shape[0] > self.T + 1:
            # For cold start, take the last T+1 frames of a long reactive path
            trimmed_traj = trajectory[-(self.T + 1):]
        elif trajectory.shape[0] == self.T + 1:
            trimmed_traj = trajectory
        else:
            return # Skip trajectories that are too short
        self.buffer.add(trimmed_traj)

    def train(self, n_iterations=100, explore_steps=64, train_steps=32, batch_size=16):
        print("Starting hybrid GFlowNet training...")
        for i in range(n_iterations):
            for _ in range(explore_steps):
                x0 = torch.tensor([-1.0], device=self.device)
                bias_fn = lambda x, t: self.bias_net(x)
                with torch.no_grad():
                    traj = self.simulator.rollout(x0, self.T, bias_fn)
                self.add_to_buffer(traj)

            if len(self.buffer) < batch_size: continue

            iter_losses = {'total': [], 'tb': [], 'score': []}
            for _ in range(train_steps):
                batch = self.buffer.sample(batch_size, device=self.device)
                if not batch: continue

                total_loss_batch, tb_loss_batch, score_loss_batch = 0, 0, 0
                for trajectory in batch:
                    loss, loss_tb, loss_score = self.loss_fn.calculate_losses(trajectory)
                    total_loss_batch += loss
                    tb_loss_batch += loss_tb
                    score_loss_batch += loss_score

                total_loss_batch /= len(batch)
                self.optimizer.zero_grad()
                total_loss_batch.backward()
                self.optimizer.step()

                iter_losses['total'].append(total_loss_batch.item())
                iter_losses['tb'].append(tb_loss_batch.item() / len(batch))
                iter_losses['score'].append(score_loss_batch.item() / len(batch))

            self.history['total'].append(np.mean(iter_losses['total']))
            self.history['tb'].append(np.mean(iter_losses['tb']))
            self.history['score'].append(np.mean(iter_losses['score']))

            if (i + 1) % 10 == 0:
                print(f"Iter {i+1}/{n_iterations} | Total: {self.history['total'][-1]:.3e} | "
                      f"TB: {self.history['tb'][-1]:.3e} | Score: {self.history['score'][-1]:.3e}")
        print("Training complete!")

# ============================================================================
# SECTION 5: EXPERIMENT & VISUALIZATION
# ============================================================================

def sample_unbiased_reactive_paths(simulator, n_paths, max_len, start_x, target_thresh):
    print(f"\nAttempting to find {n_paths} unbiased reactive paths...")
    reactive_paths = []
    max_trials = 50000
    for trial in range(max_trials):
        x0 = torch.tensor([start_x], device=simulator.device)
        with torch.no_grad():
            traj = simulator.rollout(x0, max_len)
        if traj[-1].item() > target_thresh:
            reactive_paths.append(traj.cpu())
            print(f"  > Success! Found reactive path #{len(reactive_paths)} on trial {trial+1}.")
            if len(reactive_paths) >= n_paths: break
    if not reactive_paths:
        print("  > FAILED to find any unbiased reactive paths in the given trials.")
    return reactive_paths

def plot_comparison(potential, unbiased_trajs, biased_trajs):
    fig, axes = plt.subplots(1, 2, figsize=(18, 6), gridspec_kw={'width_ratios': [2, 3]})
    x_coords = torch.linspace(-2.5, 2.5, 200).unsqueeze(-1)
    U = potential.energy(x_coords).numpy()

    axes[0].plot(x_coords.numpy(), U, 'k-', label="Potential U(x)")

    if unbiased_trajs:
        axes[0].plot(unbiased_trajs[0].numpy(), potential.energy(unbiased_trajs[0]).numpy(), 'b-', alpha=0.8, label="Unbiased MD")
    for traj in unbiased_trajs[1:]:
        axes[0].plot(traj.numpy(), potential.energy(traj).numpy(), 'b-', alpha=0.5)

    if biased_trajs:
        axes[0].plot(biased_trajs[0].numpy(), potential.energy(biased_trajs[0]).numpy(), 'r-', alpha=0.8, label="Biased GFlowNet")
    for traj in biased_trajs[1:]:
        axes[0].plot(traj.numpy(), potential.energy(traj).numpy(), 'r-', alpha=0.5)

    axes[0].set_title("Trajectories on the PES"); axes[0].set_xlabel("x"); axes[0].set_ylabel("U(x)"); axes[0].legend(); axes[0].grid(True); axes[0].set_ylim(-0.5, 10)

    if unbiased_trajs:
        axes[1].plot(unbiased_trajs[0].numpy(), 'b-', alpha=0.8, label="Unbiased MD")
    if biased_trajs:
        axes[1].plot(biased_trajs[0].numpy(), 'r-', alpha=0.8, label="Biased GFlowNet")

    axes[1].axhline(1.0, color='grey', linestyle='--', alpha=0.7); axes[1].axhline(-1.0, color='grey', linestyle='--', alpha=0.7)
    axes[1].set_title("Position vs. Time"); axes[1].set_xlabel("Time Step"); axes[1].set_ylabel("x"); axes[1].legend(); axes[1].grid(True)

    plt.suptitle("Comparison of Unbiased vs. GFlowNet-Biased Trajectories", fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95]); plt.show()

# ============================================================================
# SECTION 6: MAIN EXECUTION
# ============================================================================

if __name__ == '__main__':
    # --- 1. Setup ---
    D_unbiased = 0.3
    D_model = 0.1
    T_model = 200 # Fixed trajectory length for the model

    potential = DoubleWellPotential()
    unbiased_simulator = MDSimulator(potential, D=D_unbiased, dt=0.01, device=device)

    # --- 2. Generate Ground Truth (may take a few minutes) ---
    unbiased_reactive_paths = sample_unbiased_reactive_paths(
        simulator=unbiased_simulator, n_paths=5, max_len=15000,
        start_x=-1.0, target_thresh=0.8
    )

    # --- 3. Initialize & Train the Hybrid GFlowNet Model ---
    trainer = Trainer(potential, D=D_model, dt=0.01, T=T_model, alpha=0.01, device=device)

    if unbiased_reactive_paths:
        print(f"\nCold starting the replay buffer with {len(unbiased_reactive_paths)} ground-truth paths...")
        for traj in unbiased_reactive_paths:
            trainer.add_to_buffer(traj) # The add_to_buffer method now handles trimming
        print(f"  > Buffer size is now: {len(trainer.buffer)}")

    trainer.train(n_iterations=100)

    # --- 4. Generate Biased Paths from Trained Model ---
    print("\nGenerating biased reactive paths from trained GFlowNet...")
    biased_reactive_paths = []
    with torch.no_grad():
        for _ in range(50):
            x0 = torch.tensor([-1.0], device=trainer.device)
            traj = trainer.simulator.rollout(x0, T_model, lambda x, t: trainer.bias_net(x))
            if traj[-1].item() > 0.8:
                biased_reactive_paths.append(traj.cpu())
    print(f"  > Found {len(biased_reactive_paths)} biased reactive paths.")

    # --- 5. Visualize and Compare ---
    print("\nPlotting comparison...")
    plot_comparison(potential, unbiased_reactive_paths, biased_reactive_paths[:5])

    fig, axes = plt.subplots(1, 3, figsize=(18, 4))
    axes[0].plot(trainer.history['total']); axes[0].set_title("Total Loss"); axes[0].set_yscale('log'); axes[0].grid(True)
    axes[1].plot(trainer.history['tb']); axes[1].set_title("TB Loss"); axes[1].set_yscale('log'); axes[1].grid(True)
    axes[2].plot(trainer.history['score']); axes[2].set_title("Score Loss"); axes[2].set_yscale('log'); axes[2].grid(True)
    plt.suptitle("Training Curves (Log Scale)"); plt.show()
```

```python
# ============================================================================
# HYBRID TB–SCORE GFLOWNET - FINAL, CORRECTED & STABILIZED IMPLEMENTATION
# ============================================================================

import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.distributions import Normal
import matplotlib.pyplot as plt
from collections import deque
import random

# Use GPU if available
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Running on device: {device}")

# ============================================================================
# SECTION 1: THE PHYSICAL SYSTEM
# ============================================================================

class DoubleWellPotential:
    """Potential Energy Surface U(x) = (x^2 - 1)^2"""
    def energy(self, x): return (x**2 - 1)**2
    def force(self, x): return -4 * x * (x**2 - 1)

class MDSimulator:
    """Overdamped Langevin Dynamics Simulator"""
    def __init__(self, potential, D=0.1, dt=0.01, device='cpu'):
        self.potential, self.D, self.dt, self.device = potential, D, dt, device

    def step(self, x, bias=None, temp_factor=1.0):
        drift = self.potential.force(x) * self.dt
        if bias is not None: drift += bias * self.dt
        noise = torch.randn_like(x) * np.sqrt(2 * self.D * temp_factor * self.dt)
        return x + drift + noise

    def rollout(self, x0, T, bias_fn=None, temp_factor=1.0):
        trajectory = [x0.clone()]; x = x0.clone()
        for t in range(T):
            bias = bias_fn(x, t) if bias_fn is not None else None
            x = self.step(x, bias, temp_factor)
            trajectory.append(x.clone())
        return torch.stack(trajectory)

class OMActionCalculator:
    """Calculates Onsager-Machlup Action and its gradient (Score)"""
    def __init__(self, potential, D=0.1, dt=0.01):
        self.potential, self.D, self.dt = potential, D, dt

    def compute_action(self, trajectory):
        x_prev, x_curr = trajectory[:-1], trajectory[1:]
        force = self.potential.force(x_prev)
        residuals = x_curr - x_prev - force * self.dt
        action = (residuals**2).sum(dim=-1) / (4 * self.D * self.dt)
        return action.sum()

    def compute_score_targets(self, trajectory):
        x_prev, x_curr = trajectory[:-1], trajectory[1:]
        force = self.potential.force(x_prev)
        r = x_curr - x_prev - force * self.dt
        score_interior = (r[1:] - r[:-1]) / (2 * self.D * self.dt)
        return score_interior

# ============================================================================
# SECTION 2: THE GFLOWNET COMPONENTS
# ============================================================================

class BiasNetwork(nn.Module):
    def __init__(self, d=1, h=64): super().__init__(); self.net = nn.Sequential(nn.Linear(d, h), nn.ReLU(), nn.Linear(h, h), nn.ReLU(), nn.Linear(h, d))
    def forward(self, x): return self.net(x)

class BackwardNetwork(nn.Module):
    def __init__(self, d=1, h=64): super().__init__(); self.net = nn.Sequential(nn.Linear(d, h), nn.ReLU(), nn.Linear(h, h), nn.ReLU(), nn.Linear(h, d))
    def forward(self, x): return self.net(x)

class ScoreNetwork(nn.Module):
    def __init__(self, d=1, h=64): super().__init__(); self.net = nn.Sequential(nn.Linear(d, h), nn.ReLU(), nn.Linear(h, h), nn.ReLU(), nn.Linear(h, d))
    def forward(self, x): return self.net(x)

class ReplayBuffer:
    def __init__(self, capacity=10000):
        self.buffer = deque(maxlen=capacity)
        self.rewards = deque(maxlen=capacity)

    def add(self, trajectory, log_reward):
        self.buffer.append(trajectory.detach().cpu())
        self.rewards.append(float(log_reward))

    def sample(self, batch_size, device='cpu', gamma=0.5):
        if len(self.buffer) < batch_size: return []

        log_rewards = np.array(self.rewards)
        weights = np.exp(gamma * (log_rewards - np.max(log_rewards)))
        weights /= weights.sum()

        indices = np.random.choice(len(self.buffer), batch_size, p=weights, replace=True)
        return [self.buffer[i].to(device) for i in indices]

    def __len__(self): return len(self.buffer)

# ============================================================================
# SECTION 3: THE CORRECTED, STABILIZED HYBRID LOSS
# ============================================================================

class HybridLoss:
    def __init__(self, potential, bias_net, backward_net, score_net, D, dt, alpha, lambda_b):
        self.potential, self.bias_net, self.backward_net, self.score_net = potential, bias_net, backward_net, score_net
        self.D, self.dt, self.alpha, self.lambda_b = D, dt, alpha, lambda_b
        self.log_Z = nn.Parameter(torch.tensor(0.0))
        self.om_calc = OMActionCalculator(potential, D, dt)

    def compute_log_reward(self, trajectory, sigma):
        action = self.om_calc.compute_action(trajectory)

        dist_sq_all_points = ((trajectory - 1.0)**2).sum(dim=-1)
        similarity_all_points = torch.exp(-dist_sq_all_points / (2 * sigma**2))
        max_similarity_indicator = torch.max(similarity_all_points)
        log_indicator = torch.log(max_similarity_indicator + 1e-8)

        return -action + log_indicator

    def calculate_losses(self, trajectory, sigma_anneal):
        T = trajectory.shape[0] - 1
        x_prev, x_curr = trajectory[:-1], trajectory[1:]
        var = 2 * self.D * self.dt; std = np.sqrt(var)

        # 1. Trajectory Balance Loss (Normalized)
        force_f = self.potential.force(x_prev); bias_f = self.bias_net(x_prev)
        mean_f = x_prev + (force_f + bias_f) * self.dt
        log_pF = Normal(mean_f, std).log_prob(x_curr).sum() / T

        force_b = self.potential.force(x_curr); bias_b = self.backward_net(x_curr)
        mean_b = x_curr - (force_b + bias_b) * self.dt
        log_pB = Normal(mean_b, std).log_prob(x_prev).sum() / T

        log_R = self.compute_log_reward(trajectory, sigma=sigma_anneal) / T

        tb_error = self.log_Z + log_pF - log_R - log_pB
        loss_tb = tb_error**2

        # 2. Score Regularization Loss
        loss_score = torch.tensor(0.0, device=trajectory.device)
        if self.alpha > 0 and T > 1:
            targets = self.om_calc.compute_score_targets(trajectory)
            preds = self.score_net(trajectory[1:-1])
            if preds.shape[0] > 0: loss_score = nn.functional.mse_loss(preds, targets)

        # 3. Bias Regularization
        loss_reg = self.lambda_b * (bias_f**2).mean()

        total_loss = loss_tb + self.alpha * loss_score + loss_reg
        return total_loss, loss_tb.detach(), loss_score.detach()

# ============================================================================
# SECTION 4: THE TRAINER
# ============================================================================
class Trainer:
    def __init__(self, potential, D=0.1, dt=0.01, T=200, alpha=0.01, lambda_b=1e-4, device='cpu'):
        self.potential, self.T, self.device = potential, T, device
        self.D, self.dt = D, dt

        self.bias_net = BiasNetwork().to(device); self.backward_net = BackwardNetwork().to(device); self.score_net = ScoreNetwork().to(device)
        self.loss_fn = HybridLoss(potential, self.bias_net, self.backward_net, self.score_net, D, dt, alpha, lambda_b)
        params = (list(self.bias_net.parameters()) + list(self.backward_net.parameters()) +
                  list(self.score_net.parameters()) + [self.loss_fn.log_Z])
        self.optimizer = optim.Adam(params, lr=1e-5) # Lowered LR

        self.simulator = MDSimulator(potential, D, dt, device)
        self.buffer = ReplayBuffer()
        self.history = {'total': [], 'tb': [], 'score': []}

    def add_to_buffer(self, trajectory, log_reward):
        if trajectory.shape[0] > self.T + 1:
            trimmed_traj = trajectory[-(self.T + 1):]
        elif trajectory.shape[0] == self.T + 1:
            trimmed_traj = trajectory
        else: return
        self.buffer.add(trimmed_traj, log_reward)

    def train(self, n_iter=200, explore_steps=64, train_steps=32, batch_size=32, temp_explore=2.0):
        print("Starting stabilized hybrid GFlowNet training...")
        for i in range(n_iter):
            sigma = max(0.2, 1.5 * (0.98 ** i))

            for _ in range(explore_steps):
                x0 = torch.tensor([-1.0], device=self.device)
                with torch.no_grad():
                    traj = self.simulator.rollout(x0, self.T, lambda x, t: self.bias_net(x), temp_factor=temp_explore)
                    log_R = self.loss_fn.compute_log_reward(traj, sigma=sigma)
                self.add_to_buffer(traj, log_R)

            if len(self.buffer) < batch_size: continue

            losses = {'total': [], 'tb': [], 'score': []}
            for _ in range(train_steps):
                batch = self.buffer.sample(batch_size, self.device)
                if not batch: continue

                self.optimizer.zero_grad()
                batch_total_loss = 0
                for traj in batch:
                    loss, _, _ = self.loss_fn.calculate_losses(traj, sigma_anneal=sigma)
                    batch_total_loss += loss

                batch_total_loss /= len(batch)
                batch_total_loss.backward()
                torch.nn.utils.clip_grad_norm_(self.optimizer.param_groups[0]['params'], 5.0)
                self.optimizer.step()

                with torch.no_grad():
                    log_batch = self.buffer.sample(batch_size, self.device)
                    if not log_batch: continue
                    log_total, log_tb, log_score = 0, 0, 0
                    for traj in log_batch:
                        _, tb, sc = self.loss_fn.calculate_losses(traj, sigma_anneal=sigma)
                        log_tb += tb; log_score += sc
                    losses['tb'].append(log_tb.item() / len(log_batch))
                    losses['score'].append(log_score.item() / len(log_batch))
                    losses['total'].append(losses['tb'][-1] + self.loss_fn.alpha * losses['score'][-1])

            if losses['total']:
                for key in losses: self.history[key].append(np.mean(losses[key]))
                if (i + 1) % 10 == 0:
                    print(f"Iter {i+1}/{n_iter} | Total: {self.history['total'][-1]:.3e} | "
                          f"TB: {self.history['tb'][-1]:.3e} | Score: {self.history['score'][-1]:.3e} | Sigma: {sigma:.2f}")
        print("Training complete!")

# ============================================================================
# SECTION 5 & 6: EXPERIMENT & MAIN EXECUTION
# ============================================================================

def sample_unbiased_reactive_paths(sim, n, max_len, start_x, target):
    print(f"\nSearching for {n} unbiased reactive paths...")
    paths = []; max_trials = 50000
    for trial in range(max_trials):
        x0 = torch.tensor([start_x], device=sim.device)
        with torch.no_grad(): traj = sim.rollout(x0, max_len)
        if traj[-1].item() > target:
            paths.append(traj.cpu()); print(f"  > Found path #{len(paths)} on trial {trial+1}.")
            if len(paths) >= n: break
    if not paths: print("  > FAILED to find any unbiased paths.")
    return paths

def sample_from_trained_model(trainer, n, max_len, start_x, target):
    print(f"\nGenerating {n} biased reactive paths from trained model...")
    paths = []; max_trials = 500
    with torch.no_grad():
        for trial in range(max_trials):
            x0 = torch.tensor([start_x], device=trainer.device)
            bias_fn = lambda x, t: trainer.bias_net(x)
            traj = trainer.simulator.rollout(x0, max_len, bias_fn, temp_factor=1.0)
            if traj[-1].item() > target:
                paths.append(traj.cpu()); print(f"  > Found biased path #{len(paths)} on trial {trial+1}.")
                if len(paths) >= n: break
    print(f"  > Found {len(paths)} biased paths in {max_trials} trials.")
    return paths

def plot_comparison(potential, unbiased_trajs, biased_trajs):
    fig, axes = plt.subplots(1, 2, figsize=(18, 6), gridspec_kw={'width_ratios': [2, 3]})
    x_coords = torch.linspace(-2.5, 2.5, 200).unsqueeze(-1); U = potential.energy(x_coords).numpy()
    axes[0].plot(x_coords.numpy(), U, 'k-', label="Potential U(x)")

    if unbiased_trajs: axes[0].plot(unbiased_trajs[0].numpy(), potential.energy(unbiased_trajs[0]).numpy(), 'b-', alpha=0.8, label="Unbiased MD")
    for traj in unbiased_trajs[1:]: axes[0].plot(traj.numpy(), potential.energy(traj).numpy(), 'b-', alpha=0.5)

    if biased_trajs: axes[0].plot(biased_trajs[0].numpy(), potential.energy(biased_trajs[0]).numpy(), 'r-', alpha=0.8, label="Biased GFlowNet")
    for traj in biased_trajs[1:]: axes[0].plot(traj.numpy(), potential.energy(traj).numpy(), 'r-', alpha=0.5)

    axes[0].set_title("Trajectories on PES"); axes[0].set_xlabel("x"); axes[0].set_ylabel("U(x)"); axes[0].legend(); axes[0].grid(True); axes[0].set_ylim(-0.5, 10)

    if unbiased_trajs: axes[1].plot(unbiased_trajs[0].numpy(), 'b-', alpha=0.8, label="Unbiased MD")
    if biased_trajs: axes[1].plot(biased_trajs[0].numpy(), 'r-', alpha=0.8, label="Biased GFlowNet")

    axes[1].axhline(1.0, c='grey', ls='--', alpha=0.7); axes[1].axhline(-1.0, c='grey', ls='--', alpha=0.7)
    axes[1].set_title("Position vs. Time"); axes[1].set_xlabel("Time Step"); axes[1].set_ylabel("x"); axes[1].legend(); axes[1].grid(True)

    plt.suptitle("Comparison of Unbiased vs. GFlowNet-Biased Trajectories", fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95]); plt.show()

if __name__ == '__main__':
    # --- 1. Setup ---
    potential = DoubleWellPotential()
    unbiased_sim = MDSimulator(potential, D=0.3, dt=0.01, device=device)

    # --- 2. Generate Ground Truth ---
    unbiased_paths = sample_unbiased_reactive_paths(unbiased_sim, n=5, max_len=15000, start_x=-1.0, target=0.8)

    # --- 3. Initialize & Train with All Techniques ---
    trainer = Trainer(potential, D=0.1, dt=0.01, T=200, alpha=0.01, lambda_b=1e-5, device=device)

    if unbiased_paths:
        print(f"\nCold starting replay buffer...")
        with torch.no_grad():
            for traj in unbiased_paths:
                log_R = trainer.loss_fn.compute_log_reward(traj.to(device), sigma=1.5)
                trainer.add_to_buffer(traj, log_R)
        print(f"  > Buffer size is now: {len(trainer.buffer)}")

    # === THE FIX IS HERE ===
    # Corrected the keyword argument from n_iterations to n_iter
    trainer.train(n_iter=200, temp_explore=2.0)

    # --- 4. Generate Biased Paths ---
    biased_paths = sample_from_trained_model(trainer, n=5, max_len=2000, start_x=-1.0, target=0.8)

    # --- 5. Visualize and Compare ---
    print("\nPlotting final comparison...")
    plot_comparison(potential, unbiased_paths, biased_paths)

    fig, axes = plt.subplots(1, 3, figsize=(18, 4))
    axes[0].plot(trainer.history['total']); axes[0].set_title("Total Loss"); axes[0].set_yscale('log'); axes[0].grid(True)
    axes[1].plot(trainer.history['tb']); axes[1].set_title("TB Loss"); axes[1].set_yscale('log'); axes[1].grid(True)
    axes[2].plot(trainer.history['score']); axes[2].set_title("Score Loss"); axes[2].set_yscale('log'); axes[2].grid(True)
    plt.suptitle("Training Curves (Log Scale)"); plt.show()
```

```python
# ============================================================================
# BASELINE GFLOWNET (TB-ONLY) - FULL EXPERIMENT
# ============================================================================

import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.distributions import Normal
import matplotlib.pyplot as plt
from collections import deque
import random

# Use GPU if available
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Running on device: {device}")

# ============================================================================
# SECTION 1: THE PHYSICAL SYSTEM
# ============================================================================

class DoubleWellPotential:
    """Potential Energy Surface U(x) = (x^2 - 1)^2"""
    def energy(self, x): return (x**2 - 1)**2
    def force(self, x): return -4 * x * (x**2 - 1)

class MDSimulator:
    """Overdamped Langevin Dynamics Simulator"""
    def __init__(self, potential, D=0.1, dt=0.01, device='cpu'):
        self.potential, self.D, self.dt, self.device = potential, D, dt, device

    def step(self, x, bias=None, temp_factor=1.0):
        drift = self.potential.force(x) * self.dt
        if bias is not None: drift += bias * self.dt
        noise = torch.randn_like(x) * np.sqrt(2 * self.D * temp_factor * self.dt)
        return x + drift + noise

    def rollout(self, x0, T, bias_fn=None, temp_factor=1.0):
        trajectory = [x0.clone()]; x = x0.clone()
        for t in range(T):
            bias = bias_fn(x, t) if bias_fn is not None else None
            x = self.step(x, bias, temp_factor)
            trajectory.append(x.clone())
        return torch.stack(trajectory)

class OMActionCalculator:
    """Calculates Onsager-Machlup Action"""
    def __init__(self, potential, D=0.1, dt=0.01):
        self.potential, self.D, self.dt = potential, D, dt

    def compute_action(self, trajectory):
        x_prev, x_curr = trajectory[:-1], trajectory[1:]
        force = self.potential.force(x_prev)
        residuals = x_curr - x_prev - force * self.dt
        action = (residuals**2).sum(dim=-1) / (4 * self.D * self.dt)
        return action.sum()

# ============================================================================
# SECTION 2: THE GFLOWNET COMPONENTS
# ============================================================================

class BiasNetwork(nn.Module):
    def __init__(self, d=1, h=64): super().__init__(); self.net = nn.Sequential(nn.Linear(d, h), nn.ReLU(), nn.Linear(h, h), nn.ReLU(), nn.Linear(h, d))
    def forward(self, x): return self.net(x)

class BackwardNetwork(nn.Module):
    def __init__(self, d=1, h=64): super().__init__(); self.net = nn.Sequential(nn.Linear(d, h), nn.ReLU(), nn.Linear(h, h), nn.ReLU(), nn.Linear(h, d))
    def forward(self, x): return self.net(x)

class ReplayBuffer:
    def __init__(self, capacity=10000):
        self.buffer = deque(maxlen=capacity)
        self.rewards = deque(maxlen=capacity)

    def add(self, trajectory, log_reward):
        self.buffer.append(trajectory.detach().cpu())
        self.rewards.append(float(log_reward))

    def sample(self, batch_size, device='cpu', gamma=0.5):
        if len(self.buffer) < batch_size: return []

        log_rewards = np.array(self.rewards)
        weights = np.exp(gamma * (log_rewards - np.max(log_rewards)))
        if weights.sum() > 0:
            weights /= weights.sum()
        else: # Fallback to uniform if all weights are zero
            weights = np.ones_like(log_rewards) / len(log_rewards)

        indices = np.random.choice(len(self.buffer), batch_size, p=weights, replace=True)
        return [self.buffer[i].to(device) for i in indices]

    def __len__(self): return len(self.buffer)

# ============================================================================
# SECTION 3: TRAJECTORY BALANCE LOSS (TB ONLY)
# ============================================================================

class TrajectoryBalanceLoss:
    def __init__(self, potential, bias_net, backward_net, D, dt):
        self.potential, self.bias_net, self.backward_net = potential, bias_net, backward_net
        self.D, self.dt = D, dt
        self.log_Z = nn.Parameter(torch.tensor(0.0))
        self.om_calc = OMActionCalculator(potential, D, dt)

    def compute_log_reward(self, trajectory, sigma):
        action = self.om_calc.compute_action(trajectory)
        dist_sq_all_points = ((trajectory - 1.0)**2).sum(dim=-1)
        similarity_all_points = torch.exp(-dist_sq_all_points / (2 * sigma**2))
        max_similarity_indicator = torch.max(similarity_all_points)
        log_indicator = torch.log(max_similarity_indicator + 1e-8)
        return -action + log_indicator

    def calculate_loss(self, trajectory, sigma_anneal):
        T = trajectory.shape[0] - 1
        x_prev, x_curr = trajectory[:-1], trajectory[1:]
        var = 2 * self.D * self.dt; std = np.sqrt(var)

        force_f = self.potential.force(x_prev); bias_f = self.bias_net(x_prev)
        mean_f = x_prev + (force_f + bias_f) * self.dt
        log_pF = Normal(mean_f, std).log_prob(x_curr).sum() / T

        force_b = self.potential.force(x_curr); bias_b = self.backward_net(x_curr)
        mean_b = x_curr - (force_b + bias_b) * self.dt
        log_pB = Normal(mean_b, std).log_prob(x_prev).sum() / T

        log_R = self.compute_log_reward(trajectory, sigma=sigma_anneal) / T

        tb_error = self.log_Z + log_pF - log_R - log_pB
        return tb_error**2

# ============================================================================
# SECTION 4: THE TRAINER (TB ONLY)
# ============================================================================
class Trainer:
    def __init__(self, potential, D=0.1, dt=0.01, T=200, device='cpu'):
        self.potential, self.T, self.device = potential, T, device
        self.D, self.dt = D, dt

        self.bias_net = BiasNetwork().to(device); self.backward_net = BackwardNetwork().to(device)
        self.loss_fn = TrajectoryBalanceLoss(potential, self.bias_net, self.backward_net, D, dt)
        params = (list(self.bias_net.parameters()) + list(self.backward_net.parameters()) + [self.loss_fn.log_Z])
        self.optimizer = optim.Adam(params, lr=1e-5)

        self.simulator = MDSimulator(potential, D, dt, device)
        self.buffer = ReplayBuffer()
        self.history = {'tb_loss': []}

    def add_to_buffer(self, trajectory, log_reward):
        if trajectory.shape[0] > self.T + 1:
            trimmed_traj = trajectory[-(self.T + 1):]
        elif trajectory.shape[0] == self.T + 1:
            trimmed_traj = trajectory
        else: return
        self.buffer.add(trimmed_traj, log_reward)

    def train(self, n_iter=200, explore_steps=64, train_steps=32, batch_size=32, temp_explore=2.0):
        print("Starting GFlowNet training (TB Loss Only)...")
        for i in range(n_iter):
            sigma = max(0.2, 1.5 * (0.98 ** i))

            for _ in range(explore_steps):
                x0 = torch.tensor([-1.0], device=self.device)
                with torch.no_grad():
                    traj = self.simulator.rollout(x0, self.T, lambda x, t: self.bias_net(x), temp_factor=temp_explore)
                    log_R = self.loss_fn.compute_log_reward(traj, sigma=sigma)
                self.add_to_buffer(traj, log_R)

            if len(self.buffer) < batch_size: continue

            iter_losses = []
            for _ in range(train_steps):
                batch = self.buffer.sample(batch_size, self.device)
                if not batch: continue

                self.optimizer.zero_grad()
                batch_loss = 0
                for traj in batch:
                    batch_loss += self.loss_fn.calculate_loss(traj, sigma_anneal=sigma)

                batch_loss /= len(batch)
                batch_loss.backward()
                torch.nn.utils.clip_grad_norm_(self.optimizer.param_groups[0]['params'], 5.0)
                self.optimizer.step()
                iter_losses.append(batch_loss.item())

            if iter_losses:
                self.history['tb_loss'].append(np.mean(iter_losses))
                if (i + 1) % 10 == 0:
                    print(f"Iter {i+1}/{n_iter} | TB Loss: {self.history['tb_loss'][-1]:.3e} | Sigma: {sigma:.2f}")
        print("Training complete!")

# ============================================================================
# SECTION 5 & 6: EXPERIMENT & MAIN EXECUTION
# ============================================================================

def sample_unbiased_reactive_paths(sim, n, max_len, start_x, target):
    print(f"\nSearching for {n} unbiased reactive paths...")
    paths = []; max_trials = 50000
    for trial in range(max_trials):
        x0 = torch.tensor([start_x], device=sim.device)
        with torch.no_grad(): traj = sim.rollout(x0, max_len)
        if traj[-1].item() > target:
            paths.append(traj.cpu()); print(f"  > Found path #{len(paths)} on trial {trial+1}.")
            if len(paths) >= n: break
    if not paths: print("  > FAILED to find any unbiased paths.")
    return paths

def sample_from_trained_model(trainer, n, max_len, start_x, target):
    print(f"\nGenerating {n} biased reactive paths from trained model...")
    paths = []; max_trials = 500
    with torch.no_grad():
        for trial in range(max_trials):
            x0 = torch.tensor([start_x], device=trainer.device)
            bias_fn = lambda x, t: trainer.bias_net(x)
            traj = trainer.simulator.rollout(x0, max_len, bias_fn, temp_factor=1.0)
            if traj[-1].item() > target:
                paths.append(traj.cpu()); print(f"  > Found biased path #{len(paths)} on trial {trial+1}.")
                if len(paths) >= n: break
    print(f"  > Found {len(paths)} biased paths in {max_trials} trials.")
    return paths

def plot_comparison(potential, unbiased_trajs, biased_trajs):
    fig, axes = plt.subplots(1, 2, figsize=(18, 6), gridspec_kw={'width_ratios': [2, 3]})
    x_coords = torch.linspace(-2.5, 2.5, 200).unsqueeze(-1); U = potential.energy(x_coords).numpy()
    axes[0].plot(x_coords.numpy(), U, 'k-', label="Potential U(x)")

    if unbiased_trajs: axes[0].plot(unbiased_trajs[0].numpy(), potential.energy(unbiased_trajs[0]).numpy(), 'b-', alpha=0.8, label="Unbiased MD")
    for traj in unbiased_trajs[1:]: axes[0].plot(traj.numpy(), potential.energy(traj).numpy(), 'b-', alpha=0.5)

    if biased_trajs: axes[0].plot(biased_trajs[0].numpy(), potential.energy(biased_trajs[0]).numpy(), 'r-', alpha=0.8, label="Biased GFlowNet")
    for traj in biased_trajs[1:]: axes[0].plot(traj.numpy(), potential.energy(traj).numpy(), 'r-', alpha=0.5)

    axes[0].set_title("Trajectories on PES"); axes[0].set_xlabel("x"); axes[0].set_ylabel("U(x)"); axes[0].legend(); axes[0].grid(True); axes[0].set_ylim(-0.5, 10)

    if unbiased_trajs: axes[1].plot(unbiased_trajs[0].numpy(), 'b-', alpha=0.8, label="Unbiased MD")
    if biased_trajs: axes[1].plot(biased_trajs[0].numpy(), 'r-', alpha=0.8, label="Biased GFlowNet")

    axes[1].axhline(1.0, c='grey', ls='--', alpha=0.7); axes[1].axhline(-1.0, c='grey', ls='--', alpha=0.7)
    axes[1].set_title("Position vs. Time"); axes[1].set_xlabel("Time Step"); axes[1].set_ylabel("x"); axes[1].legend(); axes[1].grid(True)

    plt.suptitle("Comparison of Unbiased vs. GFlowNet-Biased Trajectories (TB Loss Only)", fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95]); plt.show()

if __name__ == '__main__':
    potential = DoubleWellPotential()
    unbiased_sim = MDSimulator(potential, D=0.3, dt=0.01, device=device)
    unbiased_paths = sample_unbiased_reactive_paths(unbiased_sim, n=5, max_len=15000, start_x=-1.0, target=0.8)

    trainer = Trainer(potential, D=0.1, dt=0.01, T=200, device=device)

    if unbiased_paths:
        print(f"\nCold starting replay buffer...")
        with torch.no_grad():
            for traj in unbiased_paths:
                log_R = trainer.loss_fn.compute_log_reward(traj.to(device), sigma=1.5)
                trainer.add_to_buffer(traj, log_R)
        print(f"  > Buffer size is now: {len(trainer.buffer)}")

    trainer.train(n_iter=200, temp_explore=2.0)

    biased_paths = sample_from_trained_model(trainer, n=5, max_len=2000, start_x=-1.0, target=0.8)

    print("\nPlotting final comparison...")
    plot_comparison(potential, unbiased_paths, biased_paths)

    plt.figure(); plt.plot(trainer.history['tb_loss']); plt.title("TB Loss Curve"); plt.xlabel("Iteration"); plt.ylabel("Loss"); plt.yscale('log'); plt.grid(True)
    plt.show()
```
